```python
my_dict = {}
```

```python
for url in tqdm(url_pages):
    r = requests.get(url)
    res = []
    pos = r.text.find('<a href="/title/')
    new_start = 0
    while pos != -1:
        res.append(r.text[pos + new_start + 10: pos + new_start + 25])
        new_start += pos + 30
        pos = r.text[new_start:].find('<a href="/title/')
    for i in set(res):
        url_film = 'https://www.imdb.com/{}/fullcredits?ref_=tt_ov_st_sm'.format(i)
        r_film = requests.get(url_film)
        try:
            dfs = pd.read_html(r_film.text, header=0)
            actors = dfs[2][~dfs[2].isna()['Unnamed: 1']]['Unnamed: 1'].values[:10]
            for j in actors:
                if j not in my_dict:
                    my_dict[j] = set()
                my_dict[j].add(i)
        except:
            pass
```

```python
counter = 0
for i, j in my_dict.items():
    if len(j) > 1:
        counter += 1
        print(i, j) 
print(counter)
```

```python
actors = []
for i, j in my_dict.items():
    if len(j) > 1:
        actors.append(i)
```

```python
node_labels = ['v{}'.format(i) for i in range(1, len(actors) + 1)]

act_labels = dict(zip(actors, node_labels))

vertices = pd.DataFrame({'node_label': node_labels, 'actor': actors})
vertices.head()
```

```python
vertices.to_csv('vertices.csv', index=False)
```

```python
film_dict = {}
for key, items in my_dict.items():
    if len(items) > 1:
        for item in items:
            if item not in film_dict:
                film_dict[item] = [key]
            else:
                film_dict[item].append(key)
```

```python
film_dict['title/tt4145054'], film_dict['title/tt7826108']
```

```python
actors = []
films = []
for i, j in film_dict.items():
    if len(j) > 5:
        films.append(i)
        actors.append(j)
```

```python
df_films = pd.DataFrame({'films': films, 'actors': actors})
df_films.head()
```

```python
from_nodes = []
to_nodes = []
film_labels = []

for i in range(len(df_films)):
    act_len = len(df_films['actors'][i])
    for j in range(act_len):
        for k in range(j + 1, act_len):
            from_nodes.append(act_labels[df_films['actors'][i][j]])
            to_nodes.append(act_labels[df_films['actors'][i][k]])
            film_labels.append(df_films['films'][i])
```

```python
edges = pd.DataFrame({'first_vertice': from_nodes, 'second_vertice': to_nodes, 'label': film_labels})
edges.head()
```

```python
edges.to_csv('edges.csv', index=False)
```
